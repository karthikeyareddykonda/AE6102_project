```python
# create movie from images
# image_folder/
""" https://stackoverflow.com/questions/44947505/how-to-make-a-movie-out-of-images-in-python """

import cv2
import os

DATA_SET = "data_multi_beauty"

# image_folder
video_name = f'{DATA_SET}.mp4'
image_folder = f'./pics/{DATA_SET}/'

images = [img for img in os.listdir(image_folder) if img.endswith(".png")]

print(len(images))
print(images)

frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape

# For webm, use belwo one
# fourcc = cv2.VideoWriter_fourcc('V', 'P', '0', '9')

# For mp4, use below one
fourcc = 0

FRAME_RATE = 30


video = cv2.VideoWriter(video_name, fourcc, FRAME_RATE, (width,height))

for image in images:
    print(f'{image} done')
    print(os.path.join(image_folder, image))
    video.write(cv2.imread(os.path.join(image_folder, image)))

    # Earlier, frame sizes of all images were NOT SAME. So, the video was being generated incomplete

cv2.destroyAllWindows()
video.release()
```
